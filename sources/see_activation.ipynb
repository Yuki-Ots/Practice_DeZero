# activationの観察

まずは分散1の正規分布で初期化して活性を観察する。

```python
import numpy as np
import matplotlib.pyplot as plt
import dezero
from dezero import Variable
import dezero.functions as F

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.random.randn(1000, 100)
x = Variable(x)
```

```python
node_num = 100
hidden_layer_size = 5
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
    
    w = np.random.randn(node_num, node_num) * 1.0
    
    z = F.matmul(x, w)
    a = F.sigmoid(z)
    activations[i] = a
```

```python
plt.figure(figsize=(18, 4))
for i, a in activations.items():
    a = a.data
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    plt.hist(a.flatten(), 30, range=(0, 1))
```

```python
import dezero.layers as L

x = np.random.randn(1000, 100)
x = Variable(x)
node_num = 100
hidden_layer_size = 10
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i - 1]
    layer = L.Linear(node_num, True)
    z = layer(x)
    a = F.sigmoid(z)
    activations[i] = a
    
plt.figure(figsize=(36, 4))
for i, a in activations.items():
    a = a.data
    plt.subplot(1, len(activations), i + 1)
    plt.title(str(i + 1) + "-layer")
    plt.hist(a.flatten(), 30, range=(0, 1))

plt.savefig("activations.png", dpi=600)
plt.show()
plt.figure(figsize=(4, 4))
plt.hist(activations[4].data.flatten(), 50, range=(0, 1))
plt.show()
plt.savefig("output_dist.png", dpi=600)
```

## Xavierの初期化

```python
x = np.random.randn(1000, 100)
x = Variable(x)
node_num = 100
hidden_layer_size = 5
activations = {}

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i - 1]

    w = np.random.randn(node_num, node_num) /np.sqrt(node_num)

    z = F.matmul(x, w)
    a = F.relu(z)
    activations[i] = a
    
plt.figure(figsize=(36, 4))
for i, a in activations.items():
    a = a.data
    print(x.shape)
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    a_flatten = a.flatten()
    print(np.sum(a > 0.05))
    plt.hist(a_flatten, 30, range=(0, 1))

plt.figure(figsize=(18, 4))
for i, a in activations.items():
    a = a.data
    print(x.shape)
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    b = a[a > 0.001]
    b_flatten = b.flatten()
    plt.hist(b_flatten, 30, range=(0, 1))
```
